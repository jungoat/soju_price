```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 소주 회귀 데이터 불러오기
soju_regression_data = pd.read_excel("소주 회귀.xlsx")

# 데이터 확인
print(soju_regression_data.head())

soju_regression_data.dropna(inplace=True)
# 소비자 물가지수와 참이슬클래식 가격 데이터 추출
X = soju_regression_data["소비자물가지수"].values.reshape(-1, 1)
y = soju_regression_data["참이슬클래식"].values.reshape(-1, 1)

# 선형회귀 모델 생성
model = LinearRegression()
model.fit(X, y)

# 회귀 계수 및 절편 출력
print("회귀 계수:", model.coef_[0][0])
print("절편:", model.intercept_[0])

# 회귀선 그리기
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Chamisul classic Price & Consumer Price Index linear regression')
plt.xlabel('Consumer Price Index')
plt.ylabel('Chamisul classic Price')
plt.show()
```

```python
import pandas as pd

def transpose_excel(input_file, output_file):
    # 엑셀 파일 읽기
    df = pd.read_excel(input_file)

    # 행과 열 변환
    df_transposed = df.transpose()

    # 변환된 데이터를 새 엑셀 파일로 저장
    df_transposed.to_excel(output_file, index=False)

# 예시
input_file = "정제당,유리용기.xlsx"  # 입력 파일명
output_file = "정제당_변환.xlsx"  # 출력 파일명

transpose_excel(input_file, output_file)
```
